```python
import numpy as np
import pandas as pd

print("klart")
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
tags = pd.read_csv("tags.csv")

print(movies.shape)
print(ratings.shape)
print(tags.shape)

print(movies.head())
```

```python
movies["genres_list"] = movies["genres"].apply(lambda x: x.split("|"))

print(movies[["title", "genres", "genres_list"]].head())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
genre_matrix = mlb.fit_transform(movies["genres_list"])

print(genre_matrix.shape)
print(mlb.classes_)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(movie_index, genre_matrix, movies, top_n=5):
    # ta ut vektorn 
    movie_vector = genre_matrix[movie_index].reshape(1, -1)
    
    # räkna likhet
    similarity = cosine_similarity(movie_vector, genre_matrix)[0]
    
    # sortera
    similar_indices = np.argsort(similarity)[::-1]
    
    # Ta bort sök alternativet
    similar_indices = similar_indices[1:top_n+1]
    
    return movies.iloc[similar_indices][["title", "genres"]]
```

```python
def recommend_by_title(title, movies, genre_matrix, top_n=5):
    matches = movies[movies["title"].str.contains(title, case=False)]
    
    if len(matches) == 0:
        print("Dident find any movie by that name")
        return
    
    idx = matches.index[0]
    
    print("Input film:")
    print(movies.iloc[idx][["title", "genres"]])
    print("\nRekommendationer:")
    
    return recommend(idx, genre_matrix, movies, top_n)
```
